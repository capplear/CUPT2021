```python
import numpy as np
import matplotlib.pyplot as plt
import sympy
import matplotlib as mpl
mpl.rcParams['figure.dpi']=200
```

```python
n0 = 1
n1 = 1.34
m, h, N = sympy.symbols('m, h, N')
```

```python
f1 = (m * np.pi + 2 * sympy.atan(sympy.sqrt((N**2 - n0**2) / (n1**2-N**2))))/sympy.sqrt(n1**2-N**2)
f2 = (m * np.pi + 2 * sympy.atan(n1**2 / n0**2 * sympy.sqrt((N**2 - n0**2)/(n1**2-N**2))))/sympy.sqrt(n1**2-N**2) *532 /2/np.pi
```

```python
f1 = sympy.lambdify([N,m] , f1, 'numpy')
f1 = np.vectorize(f1)
f2 = sympy.lambdify([N,m] , f2, 'numpy')
f2 = np.vectorize(f2)
```

```python
f1(1.30,0)
```

```python
plt.style.use('seaborn')
fig, ax = plt.subplots(1,1, figsize=(8,4))
N = np.linspace(n0, n1-0.01, 200)
for m in [0]:
    h1 = f1(N,m)
    h2 = f2(N, m)
    ax.plot(h1*532/2/np.pi, N, label='$m_{TE}$ = %s' % m)
    #ax.plot(h2, N, label='$m_{TM}$ = %s' % m)
ax.set_xlim(0, 10)
ax.set_ylim(1, 1.33)
ax.set_yticks([1.0, 1.1, 1.2,1.3,n1])
ax.set_xticks(532/(2*np.pi) * np.linspace(0, 10,10))

ax.set_xlabel('$h \, /\, \mathrm{nm}$',fontsize=18)
ax.set_ylabel('N',fontsize=18)
ax.axhline(1, c='black',lw=2)
ax.axhline(n1, c='black',lw=2)
```
